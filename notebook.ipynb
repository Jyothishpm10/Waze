#### Sessions 

The number of occurrences of a user opening the app during the month.

```python
#Box plot 
plt.figure(figsize = (5,1))
sns.boxplot(x = df['sessions'], fliersize = 1)
plt.title('Sessions Box Plot')
plt.show()
```

```python
#define function to plot histogram

def histogrammer(column_name, annot = True, **kwargs):
    #**kwargs any keyword arguments from sns.histplot function
    
#Histogram
    plt.figure(figsize = (5,3))
    plot = sns.histplot(x = df[column_name],**kwargs)
    median = round(df[column_name].median(),2)
    plt.axvline(median, color = 'red', ls = '--')
    if annot :
        y = plot.get_yticks()
        plt.annotate(text = f'median = {median}', xy = (median*(1+0.25), y[-3])) #annotate median value near median line
    else :
        print(f'Median : {median}')
    plt.title(f'{column_name.capitalize()} Histplot')
    plt.show()
```

```python
#plot histogram
histogrammer('sessions')
```

The sessions variable is a right-skewed distribution with half of the observations having 56 or fewer sessions. However, as indicated by the boxplot, some users have more than 700.

#### Retention by kilometer driven per driving day

```python
#describe on km_per_driving_day
df['km_per_driving_day'].describe()
```

The mean is infinity, stadard deviation is NaN and max value is ifinity. This is becuase of values zero in the column driving_days. Pandas imputes a value of infinity becuase division by zero is undefined.

Convert infinity to zero.

```python
df.loc[df['km_per_driving_day'] == np.inf, 'km_per_driving_day'] = 0
df['km_per_driving_day'].describe()
```

Consider driving 100km/hr for 12 hours a day, this will results a total travel distance of 1200 km per day. But the maximum value of 'km_per_driving_day' is 15420 to achieve this must drive above 640km/hr for 24 hours a day, which is physically impossible.

For now plot histogram by disregarding values above 1200

```python
sns.histplot(data = df, x = 'km_per_driving_day',bins = range(0,1201,20), hue = 'label', multiple = 'fill')
plt.title('Churn rate by mean km per driving day')
plt.ylabel('%', rotation = 0)
plt.show()
```

The churn rate tends to increases as the mean daily distance driven increases
